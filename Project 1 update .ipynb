# Project 1 Data

I will be working with Oregon state data for my project. 

I collected the following data sets:
    1. SARS Covid-19 cases (2020 year)
    2. Number of births in each county 
    3. Annual per capita income

Using the collected data, the following questions can be framed:
    1. Are differences in income between counties associated with differences in COVID-19 outcomes?
    2. Are differences in number of births between counties associated with differences in COVID-19 outcomes?
    3. Did counties with less than 500 COVID-19 cases have more than $50,000 annual per capita personal income and more than 800 of births?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

project_data = pd.read_csv("Main_data.csv") #Reading the .csv file

project_data
```

```python
plt.plot(project_data["Number of births"],project_data["Annual per capita personal income ($)"],'go', markersize=4)
plt.xlabel("Number of births")
plt.ylabel("Annual per capita personal income ($)")
```

```python
plt.plot(project_data["Number of Covid-19 cases"],project_data["Annual per capita personal income ($)"],'bo', markersize=4)
plt.xlabel("Number of covid-19")
plt.ylabel("Annual per capita personal income ($)")
```

```python
plt.plot(project_data["Number of Covid-19 cases"],project_data["Number of births"],'mo', markersize=4)
plt.xlabel("Number of covid-19")
plt.ylabel("Number of births)")
```

```python
fig, ax = plt.subplots(figsize =(16, 9))

ax.barh(project_data["County"], project_data["Number of Covid-19 cases"])

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
     plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
plt.xlabel("Number of Covid-19 cases in 2020")
plt.ylabel("County")
```
